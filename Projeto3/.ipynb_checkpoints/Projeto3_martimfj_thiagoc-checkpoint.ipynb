```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pandas as pd
import numpy as np

import statsmodels.api as sm1 #Usa OLS com X como matriz de planejamento
import statsmodels.formula.api as sm  #Usa OLS como no R
```

```python
dados = pd.read_excel('data.xlsx',index_col=0)
```

```python
dados.head()
```

```python
dados=dados[~np.isnan(dados.PIBPerCapta)]
dados=dados[~np.isnan(dados.Saneamento)]
dados=dados[~np.isnan(dados.CaloriasDayPerson)]
dados=dados[~np.isnan(dados.MortalidadeAbaixo5)]
dados=dados[~np.isnan(dados.IDH)]
dados.shape
```

```python
f, ax = plt.subplots(2,2,figsize=(16,16))
ax[0,0].scatter(np.log(dados.PIBPerCapta), dados.MortalidadeAbaixo5, alpha=0.7)  # Plot the raw data
ax[0,0].set_xlabel("PIB per capta")
ax[0,0].set_ylabel("Taxa de Mortalidade Infantil abaixo de 5 anos de idade (mortes/1000 nascimentos)")

ax[0,1].scatter(dados.Saneamento, dados.MortalidadeAbaixo5, alpha=0.7)  # Plot the raw data
ax[0,1].set_xlabel("% da população geral com saneamento básico")
ax[0,1].set_ylabel("Taxa de Mortalidade Infantil abaixo de 5 anos de idade (mortes/1000 nascimentos)")
ax[0,1].set_xlim(0,100)

ax[1,0].scatter(dados.CaloriasDayPerson, dados.MortalidadeAbaixo5, alpha=0.7)  # Plot the raw data
ax[1,0].set_xlabel("Calorias Consumidas diariamente por pessoa (kcal/pessoa e dia)")
ax[1,0].set_ylabel("Taxa de Mortalidade Infantil abaixo de 5 anos de idade (mortes/1000 nascimentos)")
ax[1,0].set_xlim(1500,3820)

ax[1,1].scatter(dados.IDH, dados.MortalidadeAbaixo5, alpha=0.7)  # Plot the raw data
ax[1,1].set_xlabel("Índice de Desenvolvimento Humano")
ax[1,1].set_ylabel("Taxa de Mortalidade Infantil abaixo de 5 anos de idade (mortes/1000 nascimentos)")
ax[1,1].set_xlim(0,1)

plt.tight_layout(1)
```
