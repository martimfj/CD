## Introdução

   Neste projeto, será avaliado como a educação e a entrada precoce no mercado de trabalho afetam a qualidade de vida das pessoas, até a sua aposentadoria. Presume-se que as pessoas que entram no mercado de trabalho muito cedo, trocam por consequência os estudos pelo trabalho. Isso acarreta numa má qualificação (por não ter estudado), fadando a pessoa à subempregos ou empregos de baixa remuneração. Já as pessoas que se graduam ou até fazem um doutorado/mestrado, recebem na maioria dos casos, uma remuneração alta.
   
   Essa diferença de remuneração afeta de forma supreendente a aposentadoria, visto que esse benefício está diretamente relacionado à 100% da média salarial dos últimos 10 anos de trabalho.

### Analizando pela idade, para saber se a aposentadoria é por invalidez ou por anos de contribuição/idade, vemos que a região Sudeste apresenta mais aposentados por invalidez do que a região Nordeste. E em contrapartida, a região Nordeste apresenta mais idosos aposentados.

## Conclusão 

Podemos concluir com este projeto que a educação, tida nos anos em que a população ainda é economicamente ativa (PEA) influência muito vida do aposentado, visto que quanto mais a sua qualificação, maior chances de ele ter um trabalho que remunere bem e por consequência, lhe renda uma boa aposentadoria, pelo dinheiro contribuído até se aposentar. Foi visto também, que quanto mais cedo a pessoa começa a trabalhar, mais propensa ela está a ser analfabeta, visto que ela tem uma certa chance de ter abandonado a escola para estar trabalhando.
